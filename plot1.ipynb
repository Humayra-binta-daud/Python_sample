```python
#import libraries
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import sklearn;
import seaborn as sns
```

```python
#Reading data
cos_data=pd.read_csv('tsne_csv/cosmetics.csv')
print(cos_data.head())
```

```python
#Histogram_combination
plt.figure(figsize=(6, 4))
sns.histplot(cos_data['Combination'], kde=True, bins=30, color='red')
plt.title('Histogram of cos_data Column')
plt.show()
```

```python
#Histogram_combination
plt.figure(figsize=(6, 4))
sns.histplot(cos_data['Dry'], kde=True, bins=30, color='yellow')
plt.title('Histogram of cos_data Column')
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(y=cos_data['Normal'], color='pink')
plt.title('Boxplot of Column')
plt.show()
```

```python
#All plot
sns.pairplot(cos_data, hue='Price', palette='husl')
plt.show()
```

```python
sns.pairplot(cos_data, hue='Brand', palette='husl')
plt.show()
```

```python
Cos_data_mat=cos_data[['Price','Rank']]
print(Cos_data_mat)
```

```python
Cos_data_mat=cos_data[['Price','Rank']]
cos_mat=Cos_data_mat.corr()
plt.figure(figsize=(6, 4))  # Set the figure size
sns.heatmap(cos_mat, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
